```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.metrics import  accuracy_score, recall_score, roc_auc_score, f1_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import ComplementNB, GaussianNB, CategoricalNB, BernoulliNB, MultinomialNB


```

```python
dir_path =r"C:/Users/User/Desktop/finalP/final_result/robustness_test_1.5/datasets/"
r_weekly_data=pd.read_csv(dir_path+'(raw)Quarterly_data_1.5.csv')
```

```python
#normalize data
cols_to_norm = ['future_return','volume','amount','BASIS','index_turnover_rate','short_balance_ratio','pli','spread','RSI','index_trading_volume','inventory']
r_weekly_data[cols_to_norm] = r_weekly_data[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
r_weekly_data.head()
```

```python
r_weekly_data = r_weekly_data.dropna()
r_weekly_data = r_weekly_data.drop(['quarter'], axis=1)
r_weekly_data
```

```python
X = r_weekly_data.iloc[:, :-1]  # Features (all columns except the last one)
y = r_weekly_data.iloc[:, -1]   # Labels (last column)
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
balanced_df = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
balanced_df
```

```python
# Using balanced df to make predictions
X = r_weekly_data.iloc[:, :-1]  # Features (all columns except the last one)
y = r_weekly_data.iloc[:, -1]   # Labels (last column)

X = balanced_df.iloc[:, :-1]
y = balanced_df.iloc[:, -1]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=13)
```

```python
cnb = ComplementNB().fit(X_train, y_train)  # fitting 
prediction = cnb.predict(X_test)  # predictions
prediction_train = cnb.predict(X_train)

accuracy = accuracy_score(y_test, prediction)
recall = recall_score(y_test, prediction)
roc_auc = roc_auc_score(y_test, prediction)
f1 = f1_score(y_test, prediction)
print("Accuracy:", accuracy)
print("ROC-AUC:", roc_auc)
print("F1 score:", f1)
print("Recall:", recall)
```

```python
gnb = GaussianNB().fit(X_train, y_train)
prediction = gnb.predict(X_test)  # predictions
prediction_train = gnb.predict(X_train)

accuracy = accuracy_score(y_test, prediction)
recall = recall_score(y_test, prediction)
roc_auc = roc_auc_score(y_test, prediction)
f1 = f1_score(y_test, prediction)
print("Accuracy:", accuracy)
print("ROC-AUC:", roc_auc)
print("F1 score:", f1)
print("Recall:", recall)

```

```python
clf1 = MultinomialNB()
multinb = clf1.fit(X_train, y_train)
prediction = multinb.predict(X_test)  # predictions
prediction_train = multinb .predict(X_train)

accuracy = accuracy_score(y_test, prediction)
recall = recall_score(y_test, prediction)
roc_auc = roc_auc_score(y_test, prediction)
f1 = f1_score(y_test, prediction)
print("Accuracy:", accuracy)
print("ROC-AUC:", roc_auc)
print("F1 score:", f1)
print("Recall:", recall)
```

```python
clf2 = BernoulliNB()
multinb = clf2.fit(X_train, y_train)
prediction = multinb.predict(X_test)  # predictions
prediction_train = multinb .predict(X_train)

accuracy = accuracy_score(y_test, prediction)
recall = recall_score(y_test, prediction)
roc_auc = roc_auc_score(y_test, prediction)
f1 = f1_score(y_test, prediction)
print("Accuracy:", accuracy)
print("ROC-AUC:", roc_auc)
print("F1 score:", f1)
print("Recall:", recall)
```

```python
#robustness
robust_500 = pd.read_csv(dir_path+'500_data_quarterly_1.5.csv')
#drop NaN
robust_500 = robust_500.dropna()
#drop week column
robust_500 = robust_500.drop(['q'], axis=1)
robust_500.head()
cols_to_norm = ['closingPrice','volume','amount','BASIS','turnoverRatio','short_balance_ratio','pli','spread','RSI','camount','inventory']
robust_500[cols_to_norm] = robust_500[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
X_robust = robust_500.iloc[:, :-1]  # Features (all columns except the last one)
y_robust = robust_500.iloc[:, -1]   # Labels (last column)
smote = SMOTE(random_state=42 , k_neighbors=4)
X_resampled_robust, y_resampled_robust = smote.fit_resample(X_robust, y_robust)
```

```python
# ComplementNB robustness 
prediction = cnb.predict(X_resampled_robust)  # predictions
accuracy = accuracy_score(y_resampled_robust, prediction)
recall = recall_score(y_resampled_robust, prediction)
roc_auc = roc_auc_score(y_resampled_robust, prediction)
f1 = f1_score(y_resampled_robust, prediction)
print("ComplementNB robustness")
print("Accuracy:", accuracy)
print("ROC-AUC:", roc_auc)
print("F1 score:", f1)
print("Recall:", recall)
print(" ")

# GaussianNB robustness
prediction = gnb.predict(X_resampled_robust)  # predictions
accuracy = accuracy_score(y_resampled_robust, prediction)
recall = recall_score(y_resampled_robust, prediction)
roc_auc = roc_auc_score(y_resampled_robust, prediction)
f1 = f1_score(y_resampled_robust, prediction)
print("GaussianNB robustness")
print("Accuracy:", accuracy)
print("ROC-AUC:", roc_auc)
print("F1 score:", f1)
print("Recall:", recall)
print(" ")

# MultinomialNB robustness
prediction=clf1.predict(X_resampled_robust)
accuracy = accuracy_score(y_resampled_robust, prediction)
recall = recall_score(y_resampled_robust, prediction)
roc_auc = roc_auc_score(y_resampled_robust, prediction)
f1 = f1_score(y_resampled_robust, prediction)
print("MultinomialNB robustness")
print("Accuracy:", accuracy)
print("ROC-AUC:", roc_auc)
print("F1 score:", f1)
print("Recall:", recall)
print(" ")


# BernoulliNB robustness
prediction=clf2.predict(X_resampled_robust)
accuracy = accuracy_score(y_resampled_robust, prediction)
recall = recall_score(y_resampled_robust, prediction)
roc_auc = roc_auc_score(y_resampled_robust, prediction)
f1 = f1_score(y_resampled_robust, prediction)
print("BernoulliNB robustness")
print("Accuracy:", accuracy)
print("ROC-AUC:", roc_auc)
print("F1 score:", f1)
print("Recall:", recall)
print(" ")
```
