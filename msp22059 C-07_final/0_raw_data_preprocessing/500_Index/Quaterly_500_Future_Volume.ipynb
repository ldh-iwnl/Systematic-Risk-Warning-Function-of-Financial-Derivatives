```python
import pandas as pd
volume500 = pd.read_excel("中证500股指期货成交量2022-12-31.xlsx")
```

```python
volume500
```

```python
#delete first three rows 
volume500 = volume500.iloc[3:]
#change feature name to "date" and "close"
volume500.columns = ["date", "volume"]
volume500.head()
```

```python
volume500['date']=pd.to_datetime(volume500['date'])
volume500.set_index("date", inplace=True)
q_groups=volume500.groupby(pd.Grouper(freq='Q'))
q_volume=q_groups.sum()
q_volume
```

```python
#add index week1 week2 week3...
q_volume['q'] = range(1, len(q_volume) + 1)
q_volume
```

```python
q_volume = q_volume[q_volume['volume'] != 0]
q_volume
```

```python
q_volume
```

```python
# make volume column value natural log of volume
import numpy as np
q_volume['volume'] = np.log(q_volume['volume'])
q_volume
```

```python
q_volume.set_index("q", inplace=True)
q_volume
```

```python
q_volume.to_csv("500_Future_Volume_quarterly.csv")
```
