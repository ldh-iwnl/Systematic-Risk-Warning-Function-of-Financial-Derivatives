```python
import pandas as pd
df500_closing= pd.read_excel(r"中证500股指期货收盘价(主力合约)2022-12-31.xlsx", engine='openpyxl')
df500_indexLevel= pd.read_excel(r"中证500指数2022-12-31.xlsx", engine='openpyxl')
```

```python
#drop first three row
df500_closing.drop([0,1,2], inplace=True)
df500_indexLevel.drop([0,1,2], inplace=True)
#change column name
df500_closing.columns = ['date', 'closing_price']
df500_indexLevel.columns = ['date', 'index_level']
```

```python
#combind two dataframes
df500_closing['index_level'] = df500_indexLevel['index_level']
```

```python
df500_closing['BASIS']=df500_closing['closing_price']-df500_closing['index_level']
df500_closing
```

```python
df500_closing['date']=pd.to_datetime(df500_closing['date'])
df500_closing.set_index("date", inplace=True)
q_groups=df500_closing.groupby(pd.Grouper(freq='Q'))
q_closing=q_groups.mean()
q_closing
```

```python
#add index week1 week2 week3...
q_closing['q'] = range(1, len(q_closing) + 1)
q_closing
```

```python
q_closing.set_index("q", inplace=True)
q_closing
```

```python
del q_closing['closing_price']
del q_closing['index_level']
q_closing
```

```python
#to csv
q_closing.to_csv(r'500_BASIS_quarterly.csv')
```
