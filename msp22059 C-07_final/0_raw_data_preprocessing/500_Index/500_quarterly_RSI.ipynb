```python
import pandas as pd
import numpy as np
```

```python
df_index= pd.read_excel('C:/Users/XLL/Desktop/CS毕设项目/中证500股指及股指期货数据/中证500指数2022-12-31.xlsx')
df_index.head()
```

```python
diff = np.array(df_index["index_level"])
diff = np.diff(diff)
diff
```

```python
df_index.drop([0], inplace=True)
```

```python
df_index3 = pd.DataFrame(np.insert(df_index.values,2,diff,axis = 1))
df_index3.columns = ['date','index_level','diff']
df_index3['date']=pd.to_datetime(df_index3['date'])
df_index3
```

```python
df_index3.set_index("date", inplace=True)
df_index3
```

```python
df_index4 = df_index3.drop('index_level',axis = 1)
df_index4
#check missing value
df_index4.isnull().sum()
#find out missing value
df_index4[df_index4.isnull().values==True]
df_index4.dropna(inplace=True)
df_index4
```

```python
df_index4['binary'] = df_index4['diff'].apply(lambda x: 1 if x > 0 else 0)
df_index4
```

```python
weekly_groups=df_index4.groupby([pd.Grouper(freq='Q'),'binary']).mean()
weekly_groups
```

```python
df = weekly_groups
df
```

```python
grouped = df.groupby('date')['diff']
diff1 = grouped.apply(lambda x: x.loc[x.index.get_level_values('binary') == 1].sum() / x.loc[x.index.get_level_values('binary') == 0].sum() if 0 in x.index.get_level_values('binary') and 1 in x.index.get_level_values('binary') else '+' if 1 in x.index.get_level_values('binary') else '-').to_numpy()
diff1
```

```python
diff2 = list(diff1)
diff2
```

```python
diff3 = []
for i in diff2:
    if i != '+' and i != '-':
        i = abs(float(i))
        i = 100*(1-1/(1+i))
    diff3.append(i)
diff3
```

```python
diff4 = []
for i in diff3:
    if i == '-':
        i = int('0')
    if i =='+':
        i = int('100')
    diff4.append(i)
diff4
```

```python
df_index5=df_index4.groupby([pd.Grouper(freq='Q')]).mean()
df_index5
df_index5.dropna(inplace=True)
df_index5['RSI'] = diff4
df_index5 = df_index5.drop('binary',axis = 1)
df_index5
```

```python
df_index5['quarter'] = range(1, len(df_index5) + 1)
df_index5.set_index("quarter", inplace=True)
df_index5
```

```python
df_index5.to_csv('C:/Users/XLL/Desktop/CS毕设项目/CSI300_Index & Future_0521Final/500/500_quarterly_RSI.csv', index = False, header=True)
```
