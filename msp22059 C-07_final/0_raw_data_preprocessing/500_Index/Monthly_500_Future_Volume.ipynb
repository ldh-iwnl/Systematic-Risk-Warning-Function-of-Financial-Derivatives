```python
import pandas as pd
volume500 = pd.read_excel("中证500股指期货成交量2022-12-31.xlsx")
```

```python
volume500
```

```python
#delete first three rows 
volume500 = volume500.iloc[3:]
#change feature name to "date" and "close"
volume500.columns = ["date", "volume"]
volume500.head()
```

```python
volume500['date']=pd.to_datetime(volume500['date'])
volume500.set_index("date", inplace=True)
m_groups=volume500.groupby(pd.Grouper(freq='M'))
m_volume=m_groups.sum()
m_volume
```

```python
#add index week1 week2 week3...
m_volume['m'] = range(1, len(m_volume) + 1)
m_volume
```

```python
m_volume = m_volume[m_volume['volume'] != 0]
m_volume
```

```python
m_volume
```

```python
# make volume column value natural log of volume
import numpy as np
m_volume['volume'] = np.log(m_volume['volume'])
m_volume
```

```python
m_volume.set_index("m", inplace=True)
m_volume
```

```python
m_volume.to_csv("500_Future_Volume_monthly.csv")
```
