```python
import pandas as pd
df500_futureAmount = pd.read_excel(r"中证500股指期货成交金额2022-12-31.xlsx", engine='openpyxl')
df500_futureAmount.head()
```

```python
#delete first three row
df500_futureAmount = df500_futureAmount.drop([0,1,2])
df500_futureAmount.columns = ['date', 'future_amount']
df500_futureAmount.head()
```

```python
df500_futureAmount['date']=pd.to_datetime(df500_futureAmount['date'])
df500_futureAmount.set_index("date", inplace=True)
q_groups=df500_futureAmount.groupby(pd.Grouper(freq='Q'))
q_tradingVolume=q_groups.sum()
q_tradingVolume
```

```python
#add index week1 week2 week3...
q_tradingVolume['q'] = range(1, len(q_tradingVolume) + 1)
q_tradingVolume
```

```python
# make close column value natural log of close
import numpy as np
q_tradingVolume['amount'] = np.log(q_tradingVolume['future_amount'])
q_tradingVolume
```

```python
#find out missing value
is_zero = q_tradingVolume['future_amount'] == 0
is_zero.sum()
#print out the missing value
q_tradingVolume[is_zero]
```

```python
#drop is_zero row
q_tradingVolume.drop(q_tradingVolume[is_zero].index, inplace=True)
q_tradingVolume
```

```python
q_tradingVolume.set_index("q", inplace=True)
q_tradingVolume
```

```python
del q_tradingVolume['future_amount']
#to csv
q_tradingVolume.to_csv(r'500_Future_Amount_quarterly.csv')
```
