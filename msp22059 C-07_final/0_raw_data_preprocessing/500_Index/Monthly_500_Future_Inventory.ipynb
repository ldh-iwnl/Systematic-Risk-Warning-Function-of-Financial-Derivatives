```python
import pandas as pd
inventory50 = pd.read_excel("中证500股指期货持仓量2022-12-31.xlsx")
```

```python
inventory50
```

```python
#delete first three rows 
inventory50 = inventory50.iloc[3:]
#change feature name to "date" and "close"
inventory50.columns = ["date", "inventory"]
inventory50.head()
```

```python
inventory50['date']=pd.to_datetime(inventory50['date'])
inventory50.set_index("date", inplace=True)
m_groups=inventory50.groupby(pd.Grouper(freq='M'))
m_inventory=m_groups.sum()
m_inventory
```

```python
#add index week1 week2 week3...
m_inventory['m'] = range(1, len(m_inventory) + 1)
m_inventory
```

```python
#check null
m_inventory.isnull().sum()
```

```python
m_inventory = m_inventory[m_inventory['inventory'] != 0]
m_inventory
```

```python
# make close column value natural log of close
import numpy as np
m_inventory['inventory'] = np.log(m_inventory['inventory'])
m_inventory
```

```python
m_inventory.set_index("m", inplace=True)
m_inventory
```

```python
#to csv
m_inventory.to_csv('500_monthly_inventory.csv')
```
