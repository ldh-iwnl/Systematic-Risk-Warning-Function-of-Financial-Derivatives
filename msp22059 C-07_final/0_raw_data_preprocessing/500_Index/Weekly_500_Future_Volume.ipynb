```python
import pandas as pd
volume500 = pd.read_excel("中证500股指期货成交量2022-12-31.xlsx")
```

```python
volume500
```

```python
#delete first three rows 
volume500 = volume500.iloc[3:]
#change feature name to "date" and "close"
volume500.columns = ["date", "volume"]
volume500.head()
```

```python
volume500['date']=pd.to_datetime(volume500['date'])
volume500.set_index("date", inplace=True)
weekly_groups=volume500.groupby(pd.Grouper(freq='W'))
weekly_volume=weekly_groups.sum()
weekly_volume
```

```python
#add index week1 week2 week3...
weekly_volume['week'] = range(1, len(weekly_volume) + 1)
weekly_volume
```

```python
weekly_volume = weekly_volume[weekly_volume['volume'] != 0]
weekly_volume
```

```python
# make volume column value natural log of volume
import numpy as np
weekly_volume['volume'] = np.log(weekly_volume['volume'])
weekly_volume
```

```python
weekly_volume.set_index("week", inplace=True)
weekly_volume
```

```python
weekly_volume.to_csv("Weekly_500_Future_Volume.csv")
```
