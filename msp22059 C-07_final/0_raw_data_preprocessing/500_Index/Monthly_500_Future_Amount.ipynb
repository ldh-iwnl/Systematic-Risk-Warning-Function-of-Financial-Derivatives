```python
import pandas as pd
df500_futureAmount = pd.read_excel(r"中证500股指期货成交金额2022-12-31.xlsx", engine='openpyxl')
df500_futureAmount.head()
```

```python
#delete first three row
df500_futureAmount = df500_futureAmount.drop([0,1,2])
df500_futureAmount.columns = ['date', 'future_amount']
df500_futureAmount.head()
```

```python
df500_futureAmount['date']=pd.to_datetime(df500_futureAmount['date'])
df500_futureAmount.set_index("date", inplace=True)
m_groups=df500_futureAmount.groupby(pd.Grouper(freq='M'))
m_tradingVolume=m_groups.sum()
m_tradingVolume
```

```python
#add index week1 week2 week3...
m_tradingVolume['m'] = range(1, len(m_tradingVolume) + 1)
m_tradingVolume
```

```python
# make close column value natural log of close
import numpy as np
m_tradingVolume['amount'] = np.log(m_tradingVolume['future_amount'])
m_tradingVolume
```

```python
#find out missing value
is_zero = m_tradingVolume['future_amount'] == 0
is_zero.sum()
#print out the missing value
m_tradingVolume[is_zero]
```

```python
#drop is_zero row
m_tradingVolume.drop(m_tradingVolume[is_zero].index, inplace=True)
m_tradingVolume
```

```python
m_tradingVolume.set_index("m", inplace=True)
m_tradingVolume
```

```python
del m_tradingVolume['future_amount']
#to csv
m_tradingVolume.to_csv(r'500_Future_Amount_monthly.csv')
```
