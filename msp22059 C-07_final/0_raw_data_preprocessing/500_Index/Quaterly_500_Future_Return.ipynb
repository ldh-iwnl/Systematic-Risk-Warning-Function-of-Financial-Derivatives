```python
import pandas as pd
df500_closingPrice = pd.read_excel(r"中证500股指期货收盘价(主力合约)2022-12-31.xlsx", engine='openpyxl')
```

```python
df500_closingPrice.head()
```

```python
#delete first three rows 
df500_closingPrice = df500_closingPrice.iloc[3:]
#change feature name to "date" and "close"
df500_closingPrice.columns = ["date", "close"]
df500_closingPrice.head()
```

```python
df500_closingPrice['date']=pd.to_datetime(df500_closingPrice['date'])
df500_closingPrice.set_index("date", inplace=True)
q_groups=df500_closingPrice.groupby(pd.Grouper(freq='Q'))
q_closingPrice=q_groups.mean()
q_closingPrice
```

```python
#add index week1 week2 week3...
q_closingPrice['q'] = range(1, len(q_closingPrice) + 1)
q_closingPrice
```

```python
# make close column value natural log of close
import numpy as np
q_closingPrice['close'] = np.log(q_closingPrice['close'])
q_closingPrice
```

```python
#check missing value
q_closingPrice.isnull().sum()
#find out missing value
q_closingPrice[q_closingPrice.isnull().values==True]
```

```python
#drop missing value
q_closingPrice.dropna(inplace=True)
```

```python
q_closingPrice.set_index("q", inplace=True)
q_closingPrice
```

```python
#diff function
q_closingPrice['closingPrice'] = q_closingPrice['close'].diff(periods=1)
q_closingPrice
```

```python
#delete closing_price column
del q_closingPrice['close']
q_closingPrice
```

```python
#to csv
q_closingPrice.to_csv(r'500_Future_return_quaterly.csv', index = False, header=True)
```
