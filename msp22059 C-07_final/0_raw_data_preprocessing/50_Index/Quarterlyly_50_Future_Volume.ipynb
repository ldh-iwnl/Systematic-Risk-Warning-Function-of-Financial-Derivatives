```python
import pandas as pd
volume50 = pd.read_excel("上证50股指期货成交量2022-12-31.xlsx")
```

```python
volume50
```

```python
#delete first three rows 
volume50 = volume50.iloc[3:]
#change feature name to "date" and "close"
volume50.columns = ["date", "volume"]
volume50.head()
```

```python
volume50['date']=pd.to_datetime(volume50['date'])
volume50.set_index("date", inplace=True)
q_groups=volume50.groupby(pd.Grouper(freq='Q'))
q_volume=q_groups.sum()
q_volume
```

```python
#add index week1 week2 week3...
q_volume['q'] = range(1, len(q_volume) + 1)
q_volume
```

```python
q_volume = q_volume[q_volume['volume'] != 0]
q_volume
```

```python
# make volume column value natural log of volume
import numpy as np
q_volume['volume'] = np.log(q_volume['volume'])
q_volume
```

```python
q_volume.set_index("q", inplace=True)
q_volume
```

```python
q_volume.to_csv("q_50_Future_Volume.csv")
```
