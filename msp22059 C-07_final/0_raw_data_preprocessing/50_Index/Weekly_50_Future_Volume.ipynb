```python
import pandas as pd
volume50 = pd.read_excel("上证50股指期货成交量2022-12-31.xlsx")
```

```python
volume50
```

```python
#delete first three rows 
volume50 = volume50.iloc[3:]
#change feature name to "date" and "close"
volume50.columns = ["date", "volume"]
volume50.head()
```

```python
volume50['date']=pd.to_datetime(volume50['date'])
volume50.set_index("date", inplace=True)
weekly_groups=volume50.groupby(pd.Grouper(freq='W'))
weekly_volume=weekly_groups.sum()
weekly_volume
```

```python
#add index week1 week2 week3...
weekly_volume['week'] = range(1, len(weekly_volume) + 1)
weekly_volume
```

```python
weekly_volume = weekly_volume[weekly_volume['volume'] != 0]
weekly_volume
```

```python
# make volume column value natural log of volume
import numpy as np
weekly_volume['volume'] = np.log(weekly_volume['volume'])
weekly_volume
```

```python
weekly_volume.set_index("week", inplace=True)
weekly_volume
```

```python
weekly_volume.to_csv("Weekly_50_Future_Volume.csv")
```
