```python
import pandas as pd
inventory50 = pd.read_excel("上证50股指期货持仓量2022-12-31.xlsx")
```

```python
inventory50
```

```python
#delete first three rows 
inventory50 = inventory50.iloc[3:]
#change feature name to "date" and "close"
inventory50.columns = ["date", "inventory"]
inventory50.head()
```

```python
inventory50['date']=pd.to_datetime(inventory50['date'])
inventory50.set_index("date", inplace=True)
q_groups=inventory50.groupby(pd.Grouper(freq='Q'))
q_inventory=q_groups.sum()
q_inventory
```

```python
#add index week1 week2 week3...
q_inventory['quarter'] = range(1, len(q_inventory) + 1)
q_inventory
```

```python
#check null
q_inventory.isnull().sum()
```

```python
q_inventory = q_inventory[q_inventory['inventory'] != 0]
q_inventory
```

```python
# make close column value natural log of close
import numpy as np
q_inventory['inventory'] = np.log(q_inventory['inventory'])
q_inventory
```

```python
q_inventory.set_index("quarter", inplace=True)
q_inventory
```

```python
#to csv
q_inventory.to_csv('50_quarterly_inventory.csv')
```
