```python
import pandas as pd
volume50 = pd.read_excel("上证50股指期货成交量2022-12-31.xlsx")
```

```python
volume50
```

```python
#delete first three rows 
volume50 = volume50.iloc[3:]
#change feature name to "date" and "close"
volume50.columns = ["date", "volume"]
volume50.head()
```

```python
volume50['date']=pd.to_datetime(volume50['date'])
volume50.set_index("date", inplace=True)
monthly_groups=volume50.groupby(pd.Grouper(freq='M'))
monthly_volume=monthly_groups.sum()
monthly_volume
```

```python
#add index week1 week2 week3...
monthly_volume['month'] = range(1, len(monthly_volume) + 1)
monthly_volume
```

```python
monthly_volume = monthly_volume[monthly_volume['volume'] != 0]
monthly_volume
```

```python
# make volume column value natural log of volume
import numpy as np
monthly_volume['volume'] = np.log(monthly_volume['volume'])
monthly_volume
```

```python
monthly_volume.set_index("month", inplace=True)
monthly_volume
```

```python
monthly_volume.to_csv("monthly_50_Future_Volume.csv")
```
