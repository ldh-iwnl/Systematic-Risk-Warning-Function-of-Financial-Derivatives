```python
import pandas as pd
df50_closing= pd.read_excel(r"上证50股指期货收盘价(主力合约)2022-12-31.xlsx", engine='openpyxl')
df50_indexLevel= pd.read_excel(r"上证50指数2022-12-31.xlsx", engine='openpyxl')
```

```python
#drop first three row
df50_closing.drop([0,1,2], inplace=True)
df50_indexLevel.drop([0,1,2], inplace=True)
#change column name
df50_closing.columns = ['date', 'closing_price']
df50_indexLevel.columns = ['date', 'index_level']
```

```python
#combind two dataframes
df50_closing['index_level'] = df50_indexLevel['index_level']
```

```python
df50_closing['BASIS']=df50_closing['closing_price']-df50_closing['index_level']
df50_closing
```

```python
df50_closing['date']=pd.to_datetime(df50_closing['date'])
df50_closing.set_index("date", inplace=True)
m_groups=df50_closing.groupby(pd.Grouper(freq='M'))
m_closing=m_groups.mean()
m_closing
```

```python
#add index week1 week2 week3...
m_closing['month'] = range(1, len(m_closing) + 1)
m_closing
```

```python
m_closing.set_index("month", inplace=True)
m_closing
```

```python
del m_closing['closing_price']
del m_closing['index_level']
m_closing
```

```python
#to csv
m_closing.to_csv(r'50_BASIS_monthly.csv')
```
