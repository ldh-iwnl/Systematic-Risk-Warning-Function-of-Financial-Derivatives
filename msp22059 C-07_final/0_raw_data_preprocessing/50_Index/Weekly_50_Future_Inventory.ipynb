```python
import pandas as pd
inventory50 = pd.read_excel("上证50股指期货持仓量2022-12-31.xlsx")
```

```python
inventory50
```

```python
#delete first three rows 
inventory50 = inventory50.iloc[3:]
#change feature name to "date" and "close"
inventory50.columns = ["date", "inventory"]
inventory50.head()
```

```python
inventory50['date']=pd.to_datetime(inventory50['date'])
inventory50.set_index("date", inplace=True)
weekly_groups=inventory50.groupby(pd.Grouper(freq='W'))
weekly_inventory=weekly_groups.sum()
weekly_inventory
```

```python
#add index week1 week2 week3...
weekly_inventory['week'] = range(1, len(weekly_inventory) + 1)
weekly_inventory
```

```python
#check null
weekly_inventory.isnull().sum()
```

```python
weekly_inventory = weekly_inventory[weekly_inventory['inventory'] != 0]
weekly_inventory
```

```python
# make close column value natural log of close
import numpy as np
weekly_inventory['inventory'] = np.log(weekly_inventory['inventory'])
weekly_inventory
```

```python
weekly_inventory.set_index("week", inplace=True)
weekly_inventory
```

```python
#to csv
weekly_inventory.to_csv('50_weekly_inventory.csv')
```
