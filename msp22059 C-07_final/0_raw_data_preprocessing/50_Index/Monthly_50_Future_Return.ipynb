```python
import pandas as pd
df50_closingPrice = pd.read_excel(r"上证50股指期货收盘价(主力合约)2022-12-31.xlsx", engine='openpyxl')
```

```python
df50_closingPrice.head()
```

```python
#delete first three rows 
df50_closingPrice = df50_closingPrice.iloc[3:]
#change feature name to "date" and "close"
df50_closingPrice.columns = ["date", "close"]
df50_closingPrice.head()
```

```python
df50_closingPrice['date']=pd.to_datetime(df50_closingPrice['date'])
df50_closingPrice.set_index("date", inplace=True)
m_groups=df50_closingPrice.groupby(pd.Grouper(freq='M'))
m_closingPrice=m_groups.mean()
m_closingPrice
```

```python
#add index week1 week2 week3...
m_closingPrice['month'] = range(1, len(m_closingPrice) + 1)
m_closingPrice
```

```python
# make close column value natural log of close
import numpy as np
m_closingPrice['close'] = np.log(m_closingPrice['close'])
m_closingPrice
```

```python
#check missing value
m_closingPrice.isnull().sum()
#find out missing value
m_closingPrice[m_closingPrice.isnull().values==True]
```

```python
#drop missing value
m_closingPrice.dropna(inplace=True)
```

```python
m_closingPrice.set_index("month", inplace=True)
m_closingPrice
```

```python
#diff function
m_closingPrice['closingPrice'] = m_closingPrice['close'].diff(periods=1)
m_closingPrice
```

```python
#delete closing_price column
del m_closingPrice['close']
m_closingPrice
```

```python
#to csv
m_closingPrice.to_csv(r'50_Future_return_monthly.csv', index = False, header=True)
```
