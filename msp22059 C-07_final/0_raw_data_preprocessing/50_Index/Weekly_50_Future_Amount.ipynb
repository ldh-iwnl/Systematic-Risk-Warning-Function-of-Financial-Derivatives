```python
import pandas as pd
df50_futureAmount = pd.read_excel(r"上证50股指期货成交金额2022-12-31.xlsx", engine='openpyxl')
df50_futureAmount.head()
```

```python
#delete first three row
df50_futureAmount = df50_futureAmount.drop([0,1,2])
df50_futureAmount.columns = ['date', 'future_amount']
df50_futureAmount.head()
```

```python
df50_futureAmount['date']=pd.to_datetime(df50_futureAmount['date'])
df50_futureAmount.set_index("date", inplace=True)
weekly_groups=df50_futureAmount.groupby(pd.Grouper(freq='W'))
weekly_tradingVolume=weekly_groups.sum()
weekly_tradingVolume
```

```python
#add index week1 week2 week3...
weekly_tradingVolume['week'] = range(1, len(weekly_tradingVolume) + 1)
weekly_tradingVolume
```

```python
# make close column value natural log of close
import numpy as np
weekly_tradingVolume['amount'] = np.log(weekly_tradingVolume['future_amount'])
weekly_tradingVolume
```

```python
#find out missing value
is_zero = weekly_tradingVolume['future_amount'] == 0
is_zero.sum()
#print out the missing value
weekly_tradingVolume[is_zero]
```

```python
#drop is_zero row
weekly_tradingVolume.drop(weekly_tradingVolume[is_zero].index, inplace=True)
weekly_tradingVolume
```

```python
weekly_tradingVolume.set_index("week", inplace=True)
weekly_tradingVolume
```

```python
del weekly_tradingVolume['future_amount']
#to csv
weekly_tradingVolume.to_csv(r'50_Future_Amount_weekly.csv')
```
