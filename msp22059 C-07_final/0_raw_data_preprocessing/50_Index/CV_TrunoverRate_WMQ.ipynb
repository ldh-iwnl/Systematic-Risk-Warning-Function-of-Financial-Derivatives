```python
import pandas as pd
TR = pd.read_excel(r"上证50指数换手率2022-12-31.xlsx", engine='openpyxl')
```

```python
TR.isnull().sum()
```

```python
#delete first three rows 
TR = TR.iloc[3:]
#change feature name to "date" and "close"
TR.columns = ["date", "turnoverRatio"]
TR.head()
```

```python
TR['date']= pd.to_datetime(TR['date'])
TR.set_index("date", inplace=True)
weekly_groups=TR.groupby(pd.Grouper(freq='W'))
weekly_TR=weekly_groups.mean()
weekly_TR
```

```python
#print all isnull rows
weekly_TR[weekly_TR.isnull().any(axis=1)]
```

```python
#drop na value
weekly_TR.dropna(inplace=True)
weekly_TR
```

```python
# add week column from 1 to 652 
weekly_TR['week'] = range(1, len(weekly_TR) + 1)
#set week as index
weekly_TR.set_index("week", inplace=True)
weekly_TR
```

```python
monthly_groups=TR.groupby(pd.Grouper(freq='M'))
monthly_TR=monthly_groups.mean()
monthly_TR
```

```python
#add week column from 1 to 652
monthly_TR['month'] = range(1, len(monthly_TR) + 1)
#set week as index
monthly_TR.set_index("month", inplace=True)
monthly_TR
```

```python
monthly_TR[monthly_TR.isnull().any(axis=1)]
```

```python
quarterly_groups=TR.groupby(pd.Grouper(freq='Q'))
quarterly_TR=quarterly_groups.mean()
quarterly_TR
```

```python
quarterly_TR[quarterly_TR.isnull().any(axis=1)]
```

```python
#add quarter column from 1 to 652
quarterly_TR['quarter'] = range(1, len(quarterly_TR) + 1)
#set week as index
quarterly_TR.set_index("quarter", inplace=True)
quarterly_TR
```

```python
#weekly to csv, monthly to csv
weekly_TR.to_csv(r'50_Index_TurnoverRate_Weekly.csv')
monthly_TR.to_csv(r'50_Index_TurnoverRate_Monthly.csv')
quarterly_TR.to_csv(r'50_Index_TurnoverRate_Quarterly.csv')
```
