```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.metrics import  accuracy_score, recall_score, roc_auc_score, f1_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import ComplementNB, GaussianNB, CategoricalNB, BernoulliNB, MultinomialNB


```

```python
dir_path =r"C:/Users/User/Desktop/msp22059 C-07/msp22059 C-07/2_CSI300_1.5/datasets/"
r_weekly_data=pd.read_csv(dir_path+'(raw)Monthly_data_1.5.csv')
```

```python
#normalize data
cols_to_norm = ['future_return','volume','amount','BASIS','index_turnover_rate','short_balance_ratio','pli','treasury_spread','RSI','index_trading_volume','inventory']
r_weekly_data[cols_to_norm] = r_weekly_data[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
r_weekly_data.head()
```

```python
r_weekly_data = r_weekly_data.dropna()
r_weekly_data = r_weekly_data.drop(['month'], axis=1)
r_weekly_data
```

```python
X = r_weekly_data.iloc[:, :-1]  # Features (all columns except the last one)
y = r_weekly_data.iloc[:, -1]   # Labels (last column)
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
balanced_df = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
#balanced_df.to_csv("balanced_df_weekly.csv")
balanced_df
```

```python
# Using balanced df to make predictions
X = r_weekly_data.iloc[:, :-1]  # Features (all columns except the last one)
y = r_weekly_data.iloc[:, -1]   # Labels (last column)

X = balanced_df.iloc[:, :-1]
y = balanced_df.iloc[:, -1]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100000)
```

```python
cnb = ComplementNB().fit(X_train, y_train)  # fitting 
prediction = cnb.predict(X_test)  # predictions
prediction_train = cnb.predict(X_train)

accuracy = accuracy_score(y_test, prediction)
recall = recall_score(y_test, prediction)
roc_auc = roc_auc_score(y_test, prediction)
f1 = f1_score(y_test, prediction)
print("Accuracy:", accuracy)
print("ROC-AUC:", roc_auc)
print("F1 score:", f1)
print("Recall:", recall)
```

```python
gnb = GaussianNB().fit(X_train, y_train)
prediction = gnb.predict(X_test)  # predictions
prediction_train = gnb.predict(X_train)

accuracy = accuracy_score(y_test, prediction)
recall = recall_score(y_test, prediction)
roc_auc = roc_auc_score(y_test, prediction)
f1 = f1_score(y_test, prediction)
print("Accuracy:", accuracy)
print("ROC-AUC:", roc_auc)
print("F1 score:", f1)
print("Recall:", recall)

```
